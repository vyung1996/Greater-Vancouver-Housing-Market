```python
import pandas as pd
import os
import pgeocode
```

```python
df = pd.read_excel('Zillow_Cleaned.xlsx')
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
postal_codes = df["Postal Code"].values.astype('str').tolist()
```

```python
for i in range(0,len(postal_codes)):
    postal_codes[i] = postal_codes[i].strip()
```

```python
postal_codes
```

```python
nomi = pgeocode.Nominatim('CA')
can = nomi.query_postal_code(postal_codes)
```

```python
can.to_csv('Postal_Codes',index = False)
```
